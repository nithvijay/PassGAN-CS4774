```python
import pandas as pd
import numpy as np
import os
import random
```

```python
os.listdir()
```

```python
rockyou_lines = []
with open('rockyou.txt', 'r', errors='ignore') as f:
    for line in f:
        password = line.strip()
        if len(password) < 11:
            rockyou_lines.append(password.rjust(10, '|'))
rockyou_lines
```

```python
linkedin_lines = []
with open('linkedin.txt', 'r', errors='ignore') as f:
    for line in f:
        password = line.strip().split(':')[1]
        if len(password) < 11:
            linkedin_lines.append(password.rjust(10, '|'))
linkedin_lines
```

```python
with open("rockyou_processed.txt", "w") as f:
    for password in rockyou_lines:
        f.write(password + "\n")
```

```python
with open("linkedin_processed.txt", "w") as f:
    for password in linkedin_lines:
        f.write(password + "\n")
```
